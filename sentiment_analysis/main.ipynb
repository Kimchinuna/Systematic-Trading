```python
from analysis.news_sentiment import fetch_news_headlines
from analysis.vader_sentiment import initialize_vader, analyze_sentiment_vader
```

```python
def calculate_average_sentiment(sentiment_scores):
    total_compound = sum([score['compound'] for score in sentiment_scores])
    average_sentiment = total_compound / len(sentiment_scores)
    return average_sentiment

if __name__ == "__main__":
    currency_pairs = ["USDCAD", "USDJPY", "USDEUR"]

    analyzer = initialize_vader()

    for currency_pair in currency_pairs:
        print(f"Currency Pair: {currency_pair}")
        print("-------------------------------")

        headlines = fetch_news_headlines(currency_pair)
        sentiment_scores = analyze_sentiment_vader(analyzer, headlines)

        for text, score in zip(headlines, sentiment_scores):
            print(f"Headline: {text}\nSentiment scores: {score}\n")

        average_sentiment = calculate_average_sentiment(sentiment_scores)
        print(f"Average sentiment for {currency_pair}: {average_sentiment}\n\n")
```
